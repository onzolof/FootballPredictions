```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
# read the data
data = pd.read_csv('df_clean.csv')
```

```python
data.info()
```

```python
# drop features
data_new = data.drop(["Name", "Instagram", "PositionCategory", "FormerInternational", "ActiveInternational", "InternationalTeam", "Injury", "ClubSince", "Consultancy", "Supplier", "Position", "Foot", "Nationality", "HighestValueDate", "ValueLastUpdate", "Club", "NationalLeagueLevel", "League", "LeagueCountry"], axis=1)
```

```python
for col in data_new.columns:
    sns.jointplot(data=data_new, x=data_new[col], y=data_new[col])
    plt.show()
```
