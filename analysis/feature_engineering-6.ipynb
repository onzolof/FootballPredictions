```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
# read the data
train_data = pd.read_csv('assignment.csv')
test_data = pd.read_csv('kaggle.csv')
```

```python
df = train_data.copy()
```

```python
# new features
df['car_age'] = 2022 - df['year']
df['very_old'] = df['car_age'] >= 10
df['high_mileage'] = df['mileage'] > 100_000
df['high_mpg'] = df['mpg'] > 100
df['tax_low'] = df['tax'] < 100
df['tax_medium'] = (df['tax'] >= 100) & (train_data['tax'] < 210)
df['tax_high'] = df['tax'] >= 210

# drop features
df = train_data.drop(['ID', 'tax'], axis=1)
```

```python
for col in train_data.columns:
    sns.histplot(data=train_data, x=train_data[col], bins=50)
    plt.show()
```

```python
for col in ['year', 'mileage', 'mpg', 'engineSize', 'tax']:
    sns.jointplot(data=train_data, x='price', y=col, kind="reg")
    plt.show()
```
